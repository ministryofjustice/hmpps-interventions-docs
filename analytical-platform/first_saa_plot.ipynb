Download the data mart data into a "data frame" (`df`) from the S3 bucket generated by `dbt run`:

```python
dev_derived_tables_bucket = 'mojap-derived-tables/dev/models/domain_name=interventions/'

import pyarrow.parquet as pq
import s3fs
s3 = s3fs.S3FileSystem()

df = pq.ParquetDataset(dev_derived_tables_bucket
                       + 'database_name=probation_referrals_dev_dbt/'
                       + 'table_name=referrals_first_saa_dm/'
                       , filesystem=s3).read_pandas().to_pandas()
```

Group by month (`freq='M'`) on the `index` column and use average (arithmetic `mean`) on the columns:

```python
import pandas as pd

df_idx = df.set_index('sent_date')
df_aggregate = df_idx.groupby([pd.Grouper(freq='M')]).mean()
```

Draw the graph:

```python
df_aggregate.plot()
```
